## 3. Data Cleaning

That's a data we have. So we will clean big outliers and null values

```python
# Creating boolean indexing mask based on previous researh
mask = (df['Description'].isnull()) | (df['UnitPrice'] < 0) | (df['UnitPrice'] > 5000) | (df['Quantity'] > 4000) | (df['CustomerID'].isnull())
df_clean = df.drop(df[mask].index)
df_clean = df_clean.drop_duplicates()
df_clean
```

## 4. Metrics

#### We are having not a lot of data, so we restrained by that. We will use:

Average Price

```python
df_clean['UnitPrice'].describe()
```

Avg. Purchase Per Customer

```python
df_clean['Sum'] = df['UnitPrice'] * df['Quantity']
df_clean['Sum'].loc[df_clean['Sum']>0].describe()
```

Unique Customers

```python
df_clean['CustomerID'].nunique()
```

ARPU

```python
(df_clean.loc[df_clean['Sum']>0].groupby('CustomerID')['Sum'].sum()).describe()
```

Amount of Returns

```python
df_clean.loc[df_clean['Sum']<0]['Sum'].count()
```

Percentage of returns

```python
print(round(100 * df_clean.loc[df_clean['Sum']<0]['Sum'].count() / df_clean['Sum'].count(), 2), '%')
```

Amount of coming back

```python
print(round(100 * df_clean.loc[df_clean['Sum']<0]['CustomerID'].nunique() / df_clean['CustomerID'].nunique(), 2), '%')
```

Gross Profit

```python
df_clean.loc[df_clean['Sum']>0]['Sum'].sum() - df_clean.loc[df_clean['Sum']<0]['Sum'].sum()
```

LTV

```python
repeating = df_clean['CustomerID'].duplicated(keep=False)
df_repeated = df_clean.loc[repeating]
df_customer_value = df_repeated.loc[df_repeated['Sum']>0].groupby('CustomerID')['Sum'].sum()
df_days_alive = (df_repeated.loc[df_repeated['Sum']>0].groupby('CustomerID')['InvoiceDate'].max() - df_repeated.loc[df_repeated['Sum']>0].groupby('CustomerID')['InvoiceDate'].min()).dt.days
# Because some people bought items in 1 day we cannot get a frequency from them so we are no using their data
# Their values will be equal to 0
df_days_alive = df_days_alive.loc[df_days_alive != 0]

ltv = (df_customer_value / df_days_alive).mean() * df_days_alive.mean()
ltv
```

## 5. Tops

### Top selling items

```python
fig, (ax1, ax2) = plt.subplots(1,2)
plt.subplots_adjust(wspace=0.5)


top_profit_item = df_clean.loc[df_clean['Sum']>0].groupby('StockCode')['Sum'].sum().sort_values(ascending=False).head(10)
top_profit_item.plot(kind='barh', width=0.8, ax=ax1)

top_sell_item = df_clean.loc[df_clean['Sum']>0].groupby('StockCode')['Quantity'].sum().sort_values(ascending=False).head(10)
top_sell_item.plot(kind='barh', width=0.8, ax=ax2)
```

### Sales by Month

```python
sales_daily_plot = df_clean.loc[df_clean['Sum']>0]
sales_daily_plot['InvoiceDate'] = pd.to_datetime(sales_daily_plot['InvoiceDate']).dt.to_period('M').dt.to_timestamp()
sales_daily_plot = sales_daily_plot.groupby('InvoiceDate')['Sum'].sum()

sales_daily_plot.plot(x='InvoiceDate', y='Sum',kind='line')
```

### Sales in each country

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
plt.subplots_adjust(wspace=0.8)

profit = df_clean.loc[df_clean['Sum']>0].groupby('Country')['Sum'].sum().sort_values(ascending=False)
profit.plot(kind='barh', width=0.8, align='edge', figsize=(10,8), ax=ax1)


counts = df_clean['Country'].value_counts()
counts.plot(kind='barh', width=0.8, align='edge', figsize=(10, 8), ax=ax2)

ax1.set_xscale('log')
ax2.set_xscale('log')

plt.show()
```

Most amount of transactions are in 4 countries - United Kingdom, Germant, France and ERIE. Amount of sales in UK is much more than in any other country
